#### _Lu  v cc tham s ca hm `read_csv` :_

* `encoding`: ch nh encoding ca file c vo. Mc nh l `utf-8`.
* `sep`: thay i du ngn cch gia cc ct. Mc nh l du phy `(,)`
* `header`: ch nh file c vo c header(tiu  ca cc ct) hay khng. Mc nh l `infer`.
* `index_col`: ch nh ch s ct no l ct ch s(s th t). Mc nh l `None`.
* `n_rows`: ch nh s bn ghi s c vo. Mc nh l `None`  c ton b.

```python
people_df = pd.read_csv('people.csv', encoding='utf-8', header=None, sep=',')
people_df.head(5)
```

### 6. Mt s thao tc khc trn Dataframe

_**Sp xp dataframe**_

```python
df = pd.DataFrame({'name':['Nam', 'Quang', 'Thanh', 'Hieu', 'Huyen', 'Huy'],
                    'age': [20, 20, 18, 19, 18, 20]})
print('Orginal DataFrame: \n', df)
df = df.sort_values('age', ascending=True)
print('\nDataFrame after sorting: \n', df)
```

C th sp xp theo nhiu ct c  u tin gim dn:

```python
df = df.sort_values(['age','name'], ascending=True)
print(df)
```

_**Ni 2 dataframe**__

```python
#gp 3 dataframe
df1 = pd.DataFrame({'name': ['Hiu'], 'age': [18], 'gender': ['male']})
df2 = pd.DataFrame({'name': ['Nam', 'Mai', 'Hoa'], 'age': [15,17,19]})

df = df1.append(df2, sort=False)
print(df)
df = df1.append(df2, sort=True)
print(df)
```

_**Xo trn cc bn ghi trong DataFrame**_

```python
#Xo trn cc hng trong df - shuffer dataframe r
df = pd.DataFrame({'name': ['Hiu', 'Nam', 'Mai', 'Hoa'], 'age': [18,15,17,19]})

print('Orginal DataFrame: \n', df)

df = df.sample(frac=1).reset_index(drop=True)
print('\nDataFrame after shuffling: \n', df)
```

_Gii thch thm:_

* `frac`: Ch nh s bn ghi s tr v  mi ln random. Nu bng 1, tc l random ngu nhin tt c cc bn ghi.
* `.reset_index()`: Sp xp li ct ch s ca dataframe.
* `drop`:vi gi tr True, n s ngn khng cho hm `reset_index` to ct mi t ct ch s ca dataframe ban u.
