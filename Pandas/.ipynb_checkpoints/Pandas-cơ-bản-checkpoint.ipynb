# Pandas


## 1. Giới thiệu về Pandas

Thư viện pandas trong python là một thư viện mã nguồn mở, hỗ trợ đắc lực trong thao tác dữ liệu. Đây cũng là bộ công cụ phân tích và xử lý dữ liệu mạnh mẽ của ngôn ngữ lập trình python. Thư viện này được sử dụng rộng rãi trong cả nghiên cứu lẫn phát triển các ứng dụng về khoa học dữ liệu. Thư viện này sử dụng một cấu trúc dữ liệu riêng là Dataframe. Pandas cung cấp rất nhiều chức năng xử lý và làm việc trên cấu trúc dữ liệu này. Chính sự linh hoạt và hiệu quả đã khiến cho pandas được sử dụng rộng rãi.

#### Các ứng dụng của thư viện Pandas

* DataFrame đem lại sự linh hoạt và hiệu quả trong thao tác dữ liệu và lập chỉ mục;


* Là một công cụ cho phép đọc/ ghi dữ liệu giữa bộ nhớ và nhiều định dạng file: csv, text, excel, sql database, hdf5;


* Liên kết dữ liệu thông minh, xử lý được trường hợp dữ liệu bị thiếu. Tự động đưa dữ liệu lộn xộn về dạng có cấu trúc;


* Dễ dàng thay đổi bố cục của dữ liệu;


* Tích hợp cơ chế trượt, lập chỉ mục, lấy ra tập con từ tập dữ liệu lớn.


* Có thể thêm, xóa các cột dữ liệu;


* Tập hợp hoặc thay đổi dữ liệu với group by cho phép bạn thực hiện các toán tử trên tập dữ liệu;


* Hiệu quả cao trong trộn và kết hợp các tập dữ liệu;


* Lập chỉ mục theo các chiều của dữ liệu giúp thao tác giữa dữ liệu cao chiều và dữ liệu thấp chiều;


* Tối ưu về hiệu năng;


* Pandas được sử dụng rộng rãi trong cả học thuật và thương mại. Bao gồm thống kê, thương mại, phân tích, quảng cáo,…

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

## 2. Đọc file .csv sử dụng thư viện pandas

Bạn có thể dễ dàng đọc vào một file `.csv` bằng cách sử dụng hàm `read_csv` và được trả về 1 **dataframe**. Mặc định, hàm này sẽ phân biệt các trường của file csv theo dấu phẩy

```python
people_df = pd.read_csv('people.csv')
```

Bạn có thể in ra n bản ghi đầu tiên của dataframe sử dụng hàm `head`. Ngược lại của hàm `head` là hàm `tail`

```python
people_df.head(5)
```

#### _Lưu ý về các tham số của hàm `read_csv` :_

* `encoding`: chỉ định encoding của file đọc vào. Mặc định là `utf-8`.
* `sep`: thay đổi dấu ngăn cách giữa các cột. Mặc định là dấu phẩy `(‘,’)`
* `header`: chỉ định file đọc vào có header(tiêu đề của các cột) hay không. Mặc định là `infer`.
* `index_col`: chỉ định chỉ số cột nào là cột chỉ số(số thứ tự). Mặc định là `None`.
* `n_rows`: chỉ định số bản ghi sẽ đọc vào. Mặc định là `None` – đọc toàn bộ.

```python
people_df = pd.read_csv('people.csv', encoding='utf-8', header=None, sep=',')
people_df.head(5)
```

#### c. Thêm, sửa, xóa trong dataframe

_**Thêm cột từ dữ liệu mới**_

Để thêm cột mới vào df có sắn. Trước tiên cần có 1 list dữ liệu tương ứng với số cột mà bạn muốn thêm. Thức là chiều dài của list phải tương ứng với số bản ghi của dataframe bạn muốn thêm.

Trong ví dụ này, ta sử dụng `random` để sinh ngẫu nhiên một list năm sinh và thêm vào dataframe

```python
df_len = len(people_df)
birthday = [random.randrange(1980, 2000, 1) for i in range(df_len)]
people_df['birthday'] = birthday
people_df.tail(5)
```

_**Thêm cột dựa trên dữ liệu đã có**_

Giả sử mình muốn thêm cột `is young` có giá trị `True` nếu tuổi < 25 và `False` trong trường hợp còn lại.

```python
people_df['is_young'] = people_df['age'] < 25
people_df.head(5)
```

_**Khởi tạo cột mới có giá trị bằng rỗng**_

Sử dụng cú pháp: `None`

```python
people_df['new_col'] = None
people_df[:5]
```

_**Thêm bản ghi vào dataframe**_

Cú pháp:`pandas.DataFrame.append()`

_**Sửa gía trị cột trong dataframe**_

Làm tương tự như thêm một cột mới, tên cột là tên cột bạn muốn sửa.

* `peoples_df['name'] = xxx` #list các tên mới có chiều dài bằng chiều dài của dataframe

* `peoples_df['name'] = None`// Hoặc reset trường name về None

_**Xóa cột trong dataframe**_

Cú pháp:
* Xóa một cột: `<tên_df>.drop(<tên_cột_cần_xóa>, axis=1)` 
* Xóa nhiều cột: `<tên_df>.drop(['<cột_1>', '<cột_2>'], axis=1)`  
* Xóa cột có tên:`<tên_df>.drop(column=[<tên_các_cột_cần_xóa])` 

```python
a = people_df.drop('is_young', axis=1) #dữ liệu trong people_df vẫn giữ nguyên
a.tail(5)
```

_**Xóa bản ghi theo chỉ số**_

```python
people_df.drop([0,1]).head(5) #xoa ban ghi o chi so 1 va 2
```

### 5. Tạo mới dataframe

_**Tạo mới dataframe từ python dictionary**_

```python
new_peoples = {'name': ['Nguyễn Văn An', 'Vương Vân Anh'], 'age':[20,18], 'occupations':['student', 'teacher']}
df = pd.DataFrame(new_peoples)
print(df)
```

_**Tạo mới dataframe từ python list**_

```python
name = ['Nguyen Van An', 'Vu Bang Thien']
age = [20,25]
occupation = ['student', 'enginerr']

df = pd.DataFrame()
df['name'] = name
df['age'] = age
df['occupation'] = occupation

print(df)
```

_Lưu ý: Các list phải có cùng kích thước_

### 6. Một số thao tác khác trên Dataframe

_**Sắp xếp dataframe**_

```python
df = pd.DataFrame({'name':['Nam', 'Quang', 'Thanh', 'Hieu', 'Huyen', 'Huy'],
                    'age': [20, 20, 18, 19, 18, 20]})
print('Orginal DataFrame: \n', df)
df = df.sort_values('age', ascending=True)
print('\nDataFrame after sorting: \n', df)
```

Có thể sắp xếp theo nhiều cột có độ ưu tiên giảm dần:

```python
df = df.sort_values(['age','name'], ascending=True)
print(df)
```

_**Nối 2 dataframe**__

```python
#gộp 3 dataframe
df1 = pd.DataFrame({'name': ['Hiếu'], 'age': [18], 'gender': ['male']})
df2 = pd.DataFrame({'name': ['Nam', 'Mai', 'Hoa'], 'age': [15,17,19]})

df = df1.append(df2, sort=False)
print(df)
df = df1.append(df2, sort=True)
print(df)
```

_**Xáo trộn các bản ghi trong DataFrame**_

```python
#Xáo trộn các hàng trong df - shuffer dataframe rớ
df = pd.DataFrame({'name': ['Hiếu', 'Nam', 'Mai', 'Hoa'], 'age': [18,15,17,19]})

print('Orginal DataFrame: \n', df)

df = df.sample(frac=1).reset_index(drop=True)
print('\nDataFrame after shuffling: \n', df)
```

_Giải thích thêm:_

* `frac`: Chỉ định số bản ghi sẽ trả về ở mỗi lần random. Nếu bằng 1, tức là random ngẫu nhiên tất cả các bản ghi.
* `.reset_index()`: Sắp xếp lại cột chỉ số của dataframe.
* `drop`:với giá trị True, nó sẽ ngăn không cho hàm `reset_index` tạo cột mới từ cột chỉ số của dataframe ban đầu.
