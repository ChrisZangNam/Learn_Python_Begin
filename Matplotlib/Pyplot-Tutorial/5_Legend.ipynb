# Legend


_Mục đích sử dung: Mô tả chức năng của line hoặc point, ... trong đồ thị_

**Định nghĩa hàm trong pyplot**: `legend(*args, **kwargs)`\
--> Một `legend` chứa một khóa `args` và một nhãn `kwargs`




### 1. Ví dụ tổng quan

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.gca()
ax.plot([1,2,3,4])
ax.plot([4,3,2,1])

ax.legend(['A simpke line 1-4', 'A simple line 4-1'])

plt.show()
```

Thêm một nhán vào `plot()`, giá trị sẽ được sử dungj như nhãn trong lênh `legend()`.    

Đối số duy nhất của hàm `legend()` là đối số `loc` để chỉ rõ `legend` sẽ được đặt ở đâu.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,25, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, '-b', label='sine')
plt.plot(x, y2, '-r', label='cosine')

plt.legend(loc='best') #xác định vị trí cho legend()

plt.ylim(-1.5, 2) #giới hạn cho y

plt.show()
```

#### Mã vị trí `loc` xác định của `legend`:

`best`         : 0, (only implemented for axes legends)\
`upper right`  : 1,\
`upper left`   : 2,\
`lower left`   : 3,\
`lower right`  : 4,\
`right`        : 5,\
`center left`  : 6,\
`center right` : 7,\
`lower center` : 8,\
`upper center` : 9,\
`center`       : 10,

### 2. Legend for Size of Point

```python
import pandas as pd

import io

import requests

url="https://raw.githubusercontent.com/zekelabs/data-science-ml-foundation/master/data/california_cities.csv"


#Trích xuất dữ liệu

s=requests.get(url).content

cities = pd.read_csv(io.StringIO(s.decode('utf-8')))

lat, lon = cities['latd'], cities['longd']

population, area = cities['population_total'], cities['area_total_km2']

#Vẽ các điểm rời rạc (Scatter sửu dụng size và color 
#và ko cung cấp thông tin label)
plt.scatter(lon, lat, 
            label=None,c=np.log10(population), 
            cmap='viridis',s=area, linewidth=0, alpha=0.5)

plt.axis(aspect='equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')

plt.clim(3,7)

#Tạo một legend (một danh sách tronngs với kích thước và nhãn mong muốn)
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, 
                s=area,label=str(area) + ' km$^2$')
    
plt.legend(scatterpoints=1, frameon=False, 
           labelspacing=1, title='City Area')

plt.title('California Cities: Area and Population')

plt.show()
```
