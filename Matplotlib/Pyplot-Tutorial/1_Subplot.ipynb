# Subplot Command in Matplotlib

_Chúng ta đã có nhiều ví dụ về việc vẽ đồ thị đơn trên Matplotlib. Tuy nhiên, trong nhiều trường hợp, chúng ta cần nhiều plots trong một đồ thị.
Trong bài viết này, mình sẽ hướng dẫn sử dụng lệnh *`subplot`* để thực hiện việc vẽ nhiều plots trong một đồ thị_

## 1. Giới thiệu về Subplot

   Subplot làm việc với nhiều figure và axes.
   
Cú pháp :  `subplot(nrows, ncols, plot_number)`


Trong đó:
    Một suplot được áp dụng cho một con số, con số này sẽ được chia thành các trục con `nrows` * `ncols`\
    Tham số `plot_number` xác định subplot mà hàm phải tạo ra. Có thể dao động từ 1 đến tối đa là `nrows` * `ncols`.    
    
   Nếu các giá trị của ba tham số nhỏ hơn 10, thì hàm subplot có thể được gọi với một tham số `int`, trong đó hàng trăm đại diện cho `nrows`, hàng chục đại diện cho `ncols` vàhàng đơn vị biểu diễn `plot_number`. Điều này có nghĩa là: Thay vì `subplot (2, 3, 4)` chúng ta có thể viết `subplot (234)`.

## 2. Các ví dụ minh họa

### * example 1 :

```python
import matplotlib.pyplot as plt

python_course_green = "#476042"
plt.figure(figsize=(6,4))

plt.subplot(221)#= plt.plot(2,2,1)
plt.text(0.5, 
         0.5, 
         'subplot(2,2,1)', 
         horizontalalignment='center',
         verticalalignment='center',
         fontsize=20,
         alpha=.5
        )

plt.subplot(224, axisbg=python_course_green)
plt.text(0.5, 0.5,
         'subplot(2,2,4)',
         ha='center', va='center',
         fontsize=20,
         color="y",)

plt.show()
```

### * example 2: (Cách khác của example 1)

```python
import matplotlib.pyplot as plt

python_course_green = '#476042'

fig = plt.figure(figsize=(6,4))

sub1 = fig.add_subplot(121)
sub1.text(0.5, 0.5,
          'subplot(1,2,1)',
          horizontalalignment='center',
          verticalalignment='center',
          fontsize=20,
          alpha=.5)

sub2 = fig.add_subplot(122, axisbg=python_course_green)
sub2.text(0.5, 0.5,
          'subplot(1,2,2)',
          ha='center', va='center',
          color='y')

plt.show()
```

Để loại bỏ hoặc tùy biến ticks trong các `subplot`, với cách tiếp cận đầu tiên, ta có thể sử dụng phuong thức `xticks`, `yticks` của pyplot.

Tuy nhiên, ta có thể sử dụng phương thức `set_xticks()`, `set_yticks()` của lớp `axes` để trả về từ phương thức `add_subplot()`.


### _Các ví dụ về chia các subplot:_

```python
import matplotlib.pyplot as plt

X = [(2,1,1), (2,3,4), (2,3,5), (2,3,6)]

for nrows, ncols, plot_number in X:
    plt.subplot(nrows, ncols, plot_number)
    
plt.show()
```

```python
import matplotlib.pyplot as plt

X = [(2,1,2), (2,3,1), (2,3,2), (2,3,3)]

for nrows, ncols, plot_number in X:
    plt.subplot(nrows, ncols, plot_number)
    
plt.show()
```

```python
import matplotlib.pyplot as plt

X = [(4,2,1), (4,2,2), (4,2,3), (4,2,5),(4,2,(4,6)), (4,1,4)]

for nrows, ncols, plot_number in X:
    plt.subplot(nrows, ncols, plot_number)
    
plt.show()
```
